```python
import os
from lameg.surf import postprocess_freesurfer_surfaces

SUB = ['105', '107', '109', '113', '114',
           '115', '117','118','119', '123', 
           '124', '126','127','129', '130',
           '131','132','133','134','135', 
           '136', '137','138', '141','142',
           '143','144','145']

data_path = ""

for s, sub_n in enumerate(SUB):
    
    surf_output_dir = os.path.join(data_path, f"{sub_n}-synth", "layer_surf")
    os.makedirs(surf_output_dir, exist_ok=True)

    postprocess_freesurfer_surfaces(
    f"{sub_n}-synth",
    surf_output_dir,
    f"sub-{sub_n}-multilayer.11.ds.link_vector.fixed.gii",
    n_surfaces=11,
    ds_factor=0.1,
    orientation='link_vector',
    remove_deep=True
    )
    
    print(f"Multilayer mesh for {sub_n} saved in {surf_output_dir}.")
```
